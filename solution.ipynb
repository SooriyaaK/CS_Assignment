```python
import csv
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Read the CSV file
csv_data = pd.read_csv("istherecorrelation.csv", delimiter=';')

# View the first 5 rows
csv_data
```

```python
csv_data["WO [x1000]"] = csv_data["WO [x1000]"].str.replace(',', '.')
csv_data["WO [x1000]"] = csv_data["WO [x1000]"].astype(float)
```

```python
fig, ax1 = plt.subplots()

scatterplot = plt.scatter(
    csv_data["NL Beer consumption [x1000 hectoliter]"],
    csv_data["Year"],
    c=csv_data["WO [x1000]"],
    cmap="viridis",
    
)
plt.xlabel("NL Beer consumption [x1000 hectoliter]")
plt.ylabel("Year")
plt.legend()


cbar = plt.colorbar(scatterplot)
label = cbar.set_label("WO [x1000]")
plt.title('Trends in  Dutch Beer consumption (2006 - 2018) by WO')
plt.savefig("plot.png", dpi=300)
plt.show()
```
